## Exercise 10: Genetic algorithms and Parallel computing

In the following exercise the performance of the genetic algorithm developed in exercise 09 is bettered through the use of parallel computing. The use of parallel computing is made possible via the MPI library in C++

### Brief overview of parallel computing and implementation

Parallel computing makes it possible to perform simultanaeous calculations on one computer having more than one *core*. This possibility can be used in order to make performances better in some problems. In the following exercise the parallelization of the algorithm works as follows:
- the main code is executed simulatenously by 4 cores
- every N iterations of the genetic algorithm 2 cores (randomly chosen) exchange their best individual. N is chosen $N = 100$ for both cases.

### Exercise 10.01/02: 34 random cities and American Capitals

The actual parallelized genetic algorithm is applied to 2 cases:
- 34 cities randomly placed in a square of side 1 (same as exercise 09)
- the 50 American capitals

The parameters of the exercise for the mutations are the same as in exercise 09. The varied parameters are the following:
- 34 square cities: $M = 400$ chromosomes $N = 4\times 10^4$ iterations of the GA which means, counting a "generation" every $400$ iterations, $100$ generations 
- American: $M = 1000$ chromosomes $N = 10^5$ iterations of the GA which means, counting a "generation" every $1000$ iterations, $100$ generations 

```python
fit = []
fit.append(np.loadtxt('Square/Mean_best_half_0.txt'))
fit.append(np.loadtxt('Square/Mean_best_half_1.txt'))
fit.append(np.loadtxt('Square/Mean_best_half_2.txt'))
fit.append(np.loadtxt('Square/Mean_best_half_3.txt'))
x = np.arange(0,np.size(fit[0]),1)
```

```python
fig, ax = plt.subplots(figsize=(15,10))
for i in range(4):
    ax.plot(x,fit[i], label = ('Best half mean of rank',i))
ax.set_title('Best half fitness for cities randomly placed in square', fontsize = 15)
ax.legend()
ax.set_xlabel('Generation')
ax.set_ylabel('Path length')
ax.grid(True)
plt.show()
```

Already from this graph it can be noticed that the GA performs better when parallelized as it takes less time to converge (roughly a quarter of the unparallelized case, cfr. Exercise 09). 

```python
pop = []
pop.append(np.loadtxt("Square/Best_individual_0.txt"))
pop.append(np.loadtxt("Square/Best_individual_1.txt"))
pop.append(np.loadtxt("Square/Best_individual_2.txt"))
pop.append(np.loadtxt("Square/Best_individual_3.txt"))
cities_x, cities_y = np.loadtxt('Square/square_cities.dat', usecols = (0,1), unpack = True)
```

```python
fig,ax = plt.subplots(2,2,figsize = (15,15))
fig.tight_layout(pad = 3.0)
x_0 = np.zeros((4,2,np.size(cities_x)+1))
for j in range(4):
    i=0
    for city in pop[j]:
        x_0[j][0][i] = cities_x[int(city)]
        x_0[j][1][i] = cities_y[int(city)]
        i+=1
    x_0[j][0][np.size(cities_x)] = x_0[j][0][0]
    x_0[j][1][np.size(cities_x)] = x_0[j][1][0]
    ax[int(j/2),j%2].scatter(x_0[j][0],x_0[j][1], marker='o', color = 'black')
    ax[int(j/2),j%2].scatter(x_0[j][0][0],x_0[j][1][0], marker='*', s=250, color='green')
    ax[int(j/2),j%2].plot(x_0[j][0],x_0[j][1], color='red', linewidth=1)
    ax[int(j/2),j%2].set_xlabel('$x$')
    ax[int(j/2),j%2].set_ylabel('$y$')
    ax[int(j/2),j%2].set_title('Rank %i'% j, fontsize = 10)
    ax[int(j/2),j%2].grid()
```
