## Exercise 08: Variational Monte-Carlo for QM ground state problem

The goal of the following exercise is to find an approximate ground state wave function $\Psi(x)$ for a 1D particle. The Hamiltonian of the system is:
$$
\mathcal{H} = K + x^4 -\frac{5}{2} x^2
$$

where $K$ is the kinetic energy operator that in position basis becomes $K = \frac{-\hbar^2}{2m}\frac{\partial^2}{\partial x^2}$
This problem is not analitically solvable so it needs to be attacked with computational tools. The starting point of the problem is to have a trial wavefunction possibly depending on parameters (which can then in principle be varied in order to get the best possible result). The choice for this problem is:

$$
\Psi_{\mu,\sigma}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+e^{-\frac{(x+\mu)^2}{2\sigma^2}}
$$

### Exercise 08.1: Sampling of probability distribution

As is known from the basic principles of quantum mechanics the modulus squared of a wavefunciton has the meaning of a probability density function. the calculation of mean energy in a given quantum state is given by:

$$
\langle \mathcal{H}\rangle_{\Psi} = \frac{\int dx \Psi^*(x) {\hat H} \Psi(x)}
{\int dx |\Psi(x)|^2} = 
\int dx \frac{|\Psi(x)|^2}{\int dx |\Psi(x)|^2} \frac{{\hat H} \Psi(x)}{\Psi(x)}
$$

clearly the probability density funciton in represented by $\int dx \frac{|\Psi(x)|^2}{\int dx |\Psi(x)|^2}$ and this can be sampled by means of algorithm already seen. In this case the Metropolis is used with a uniform transition probability $T(x_{new}|x_{old})$. For this first part of the exercise the *values of the parameters are fixed*; below are both shown the plots for the sampling and for the calculation of energy with data blocking (number of steps $N = 10^6$; number of blocks $M = 100$).
The choice for the parameters are:

### Exercise 08.2: simulated annealing for minimization problem

In the second part of the exercise the goal is to find the best values for the two parameters entring the choice of the trial wavefunction. This goal is reached with the use of simulated annealing. In a similair fashion fo the case of the Ising model the annealing process works in this way:
- a starting temperature and the starting values of the parameters are chosen
- the energy for the starting values is computed
- a move varying the values of the parameters is proposed
- the new energy is computed 
- the move is proposed with probability given by: $p = min(1,e^{-\beta(E_{new}-E_{old})})$
- the temperature is then lowered (the specific way and times in which the temperature is lowered have to be chosen)

For this code the followed instructions were:

- $\beta_{0} = 0.5$; $\mu_{0} = 1$; $\sigma_{0} = 0.5$
- for each $\beta_i$ the number $n_i$ of sampling is given by the rule: $n_i = 10+2int(i/100)$; this means that when the temperature is lowered further the system is given more time to explore the "surroundings" in parameters space
- for $E > 0.4 \to \delta \mu = 0.1 \text{ and } \delta \sigma = 0.1$ 
- for $E \in [-0.4,0.4] \to \delta \mu = 0.01 \text{ and } \delta \sigma = 0.005$
- for $E < -0.4 \to \delta \mu = 0.005 \text{ and } \delta \sigma = 0.001$
- $\beta_i =\beta_{i-1}+1$ 

The rules about the energy were given because of a previous knowledge of the minimum value of the energy computed in another way.

```python
def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

print("Ground state energy: ", E[0])
print("1st excited state energy: ", E[1])
print("2nd excited state energy: ", E[2])
```
