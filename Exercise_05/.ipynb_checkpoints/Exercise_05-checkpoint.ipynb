## Exercise 05: Monte Carlo/Metropolis integration for hydrogen atom

In quantum mechanics the hydrogen atom energy eigenfunctions are explicitly solved via the use of Laguerre polynomials. Once the form of an eigenfunction is known quantities of interest such as the mean value of observables can be computed via computational tools. In this exercise the Metropolis algorithm has been used.

### Brief overview on Markov chains

Let's say a system can be in a discrete set of states $X_i$ with $i = 1,...,n$. A discrete Markov chain can be represented via a matrix $T_ij$ whose entries mean $T_{ij} = \text{probability for system to go from state } i\text{ to state } j$ or via a graph (see image below). 
<img src="MC.png" alt="drawing" width="200"/>
Without getting into too much detail the most important propoerty of a transition matrix is for it to be *regular*.

A transition matrix is regular if there exists a number $m>0$ such that $(T^m)_{ij} > 0 \; \forall i,j$. Regularity is a useful property because it makes possible the application of Markov's theorem.

If a transition matrix $T$ is regular it admits a unique invariant distribution $p^*$ and for any initial state $\pi^{(0)}$ the following holds:

$$
p^*_i = \lim_{k\to\infty}\left(\pi^{(0)}T^k\right)_i
$$

How is this related to the problem of sampling a target distribution? Somehow if the goal is to sample a distribution $p^*$ one could be able to find a related Markov process for which $p^*$ is the invariant distribution. It is now that the Metropolis algorithm comes into play.

### Metropolis sampling

The Metropolis sampling is an accept/reject technique; given a target distribution that is wished to be sampled a particular kind of Random Walk is initiated. Suppose $x_1,x_2,..$ are the steps in the RW and that each of the $x_i$ is a random variable associated to a probability distribution $d_i$. What is wished is that the $d_i$ tend in the limit to the target distribution $p$:

$$
\lim_{i\to\infty}d_i = p
$$

To build such a random walk the following observations are needed.
Let's consider a stationary Markov process defind as:
$$
p_1(x,t+\Delta t) = p_1(x,t) = p(x)
$$

$$
p_{1|1}(x_2,t_2|x_1,t_1) = p_t(x_2|x_1) \, \text{with } t = t_2-t_1
$$

Considering a discrete sequence of equally spaced times $t_i$ the conditional probability above defined depends only on the events $x_n,x_{n+1}$ 

$$
p_{1|1}(x_2,t_2|x_1,t_1) = p_t(x_2|x_1)=K(x_2|x_1)
$$

yielding

$$
p(x) = \int K(x|y)p(y)dy
$$

Finally a RW $d_i$ can be built using the found *stochastic nucleus* $K$ as:

$$
d_{n+1} = \int K(x|y)d_n(y)dy
$$

Skipping a few mathematical aspects, the final goal is to find a suitable stochastic nucleus in order to build the RW $d_{i}$. $K$ may is decomposed in a trial transition probability $T(x|y)$ times a probability to accept the proposed move $A(x|y)$ (this gives useful properties to the nucleus $K$) 

$$
K(x|y) = T(x|y)A(x|y)
$$

$T(x|y)$ is the one chosen between for example a uniform distribution of a Gaussian distribution etc.; $A(x|y)$ meanwhile is chosen to be 

$$A(x|y) = \min[1,q(x|y)], \, \text{where } q(x|y) = \frac{T(y|x)p(x)}{T(x|y)p(y)}\geq 0$$

A noticeable feature of Metropolis sampling is that the normalization of the PDF is not needed in order to sample it.

Clearly the results can be heavily dependent on the choice of the proposed moves encoded in the choice of $T(x|y)$. Any kind of chosen $T$ will depend on few parameters and the empirical rule followed is to tune these parametrs in order to have an *accepted/total* ratio around $50\%$. 

### Exercise 05.1: Metropolis sampling

The goal of the exercise is to compute the expectation value of position operator in two different states:
$$
\langle r \rangle_{\Psi_{1,0,0}} = \frac{3}{2}a_0 \,\,\, \langle r \rangle_{\Psi_{2,1,0}} = 5 a_0
$$
where the two states are explicitly given by:
$$
\Psi_{1,0,0}(r,\theta,\phi)=
\frac{a_0^{-3/2}}{\sqrt{\pi}} e^{-r/a_0} 
$$
and 
$$
\Psi_{2,1,0}(r,\theta,\phi)=
\frac{a_0^{-5/2}}{8}\sqrt{\frac{2}{\pi}} r e^{-r/2a_0} \cos(\theta)
$$
From quantum mechanics it is known that $|\Psi(x,y,z)|^2$ represents a probability density function; since the expectation value of an observable (depending on position only) $\hat{\mathcal{O}}(x,y,z)$ reduces to an integral of the form:

$$
\frac{\int_{\mathbb{R}^3}d^3 r\Psi^*(\vec{r})\mathcal{O}(\vec{r})\Psi(\vec{r})}{\int_{\mathbb{R}^3}d^3 r\Psi^*(\vec{r})\Psi(\vec{r})} = \frac{\int_{\mathbb{R}^3}d^3 r|\Psi(\vec{r})|^2\mathcal{O}(\vec{r})}{\int_{\mathbb{R}^3}d^3 r|\Psi(\vec{r})|^2}
$$

the problem reduces to the sampling of the distribution $|\Psi(\vec{r})|^2$; this goal is reached through the use of Metropolis sampling using a uniform and Gaussian distributed proposed moves. For reference the parameters of the proposed moves are:

- widht of uniform sampling for GS $=  2.5$
- widht of uniform sampling for 1st excited state $=  4.5$
- $\sigma$ of gaussian sampling for GS $=  0.7$ ($\mu = 0$) 
- $\sigma$ of gaussian sampling for 1st excited state $=  2.0$ ($\mu = 0$)

The starting point of the simulation is $\vec{x}_0 = (0,0,1)$

The results are plotted as usual with data blocking with $10^6$ moves divided in $100$ blocks. An equilibration plot is also shown with $500$ moves before starting the actual simulation 

```python
a1, b1 = np.loadtxt("Equilibration_unif_1.txt", usecols=(0,1), unpack='true')
a2, b2 = np.loadtxt("Equilibration_unif_2.txt", usecols=(0,1), unpack='true')
a3, b3 = np.loadtxt("Equilibration_gauss_1.txt", usecols=(0,1), unpack='true')
a4, b4 = np.loadtxt("Equilibration_gauss_2.txt", usecols=(0,1), unpack='true')

x1, f1, error1, exp1 = np.loadtxt("Unif_int_eval_1.txt", usecols=(0,1,2,3), unpack='true')
x2, f2, error2, exp2 = np.loadtxt("Unif_int_eval_2.txt", usecols=(0,1,2,3), unpack='true')
x3, f3, error3, exp3 = np.loadtxt("Gauss_int_eval_1.txt", usecols=(0,1,2,3), unpack='true')
x4, f4, error4, exp4 = np.loadtxt("Gauss_int_eval_2.txt", usecols=(0,1,2,3), unpack='true')
```

```python
fig, axs= plt.subplots(4,2, figsize=(20,30))
fig.tight_layout(pad = 5.0)


axs[0][0].plot(a1, b1)
axs[0][0].set_title("Equilibration of uniform sampling GS", fontsize=20)
axs[1][0].plot(a2, b2)
axs[1][0].set_title("Equilibration of uniform sampling 1st excited state", fontsize=20)
axs[2][0].plot(a3, b3)
axs[2][0].set_title("Equilibration of Gaussian sampling GS", fontsize=20)
axs[3][0].plot(a4, b4)
axs[3][0].set_title("Equilibration of Gaussian sampling 1st excited state", fontsize=20)
axs[0][1].errorbar(x1,f1,yerr=error1)
axs[1][1].errorbar(x2,f2,yerr=error2)
axs[2][1].errorbar(x3,f3,yerr=error3)
axs[3][1].errorbar(x4,f4,yerr=error4)
axs[0][1].plot(x1, exp1, label = 'expected value')
axs[1][1].plot(x2, exp2, label = 'expected value')
axs[2][1].plot(x3, exp3, label = 'expected value')
axs[3][1].plot(x4, exp4, label = 'expected value')
axs[0][1].set_title("Expectation value of distance uniform sampling GS", fontsize=20)
axs[1][1].set_title("Expectation value of distance uniform sampling 1st excited state", fontsize=20)
axs[2][1].set_title("Expectation value of distance Gaussian sampling GS", fontsize=20)
axs[3][1].set_title("Expectation value of distance Gaussian sampling 1st excited state", fontsize=20)
for i in range(4): 
    axs[i][0].set_xlabel("time", fontsize=20)
    axs[i][1].set_xlabel("block number", fontsize=20)
    axs[i][0].set_ylabel(r'$\langle r \rangle$', fontsize=20)
    axs[i][1].set_ylabel(r'$\langle r \rangle$', fontsize=20)
    axs[i][1].legend()

plt.show()
```
